```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import joblib
```

```python
df = pd.read_csv('Data .csv')
print(df)
```

```python
x = df[['Path', 'Identity', 'Expression']]
y = df[['Age']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import os
import skimage
from skimage import io
from skimage import transform

# Specify the path to the folder containing your images
folder_path = '/content/drive/MyDrive/Data_resize_train'

# List all files in the folder
file_names = os.listdir(folder_path)

lst = []
for file_name in file_names:
    try:
        # Construct the full file path by joining the folder path with the file name
        file_path = os.path.join(folder_path, file_name)

        # Read the image using io.imread
        curr_image = io.imread(file_path)

        # Process the image (resize, flatten, normalize) as needed
        img_resize = transform.resize(curr_image, output_shape=(32, 32, 3))
        img_reshape = np.reshape(img_resize, (32 * 32 * 3))
        normalized_img = img_reshape / 255.0

        lst.append(normalized_img)
    except Exception as e:
        print(f"Error processing {file_name}: {e}")
        # Handle the error, for example, skip the problematic file and continue processing the next file
        # continue  # Uncomment this line if you want to skip the problematic file and continue processing the next file

x_train = np.array(lst)
x_train
```

```python
y_train= np.array(y_train)
y_train
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```
