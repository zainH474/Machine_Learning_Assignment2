```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import joblib
```

```python
df = pd.read_csv('Data.csv')
print(df)
```

```python
x = df[['Path', 'Identity', 'Expression']]
y = df[['Age']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import skimage
from skimage import io
from skimage import transform

#Specifying the path
path = x_train['Path']

arr = []
for files in path:
    image = io.imread(files)
    #resizing the image to match a certain type
    img_resize = transform.resize(image, output_shape=(32,32,3))
    img_reshape = np.reshape(img_resize, (32*32*3))
    arr.append(img_reshape/255.)

x_train = np.array(arr)
print(x_train)
```

```python
#Now do the same for test data
#No need to import since it has been done above in google colab

#Specifying the path
path = x_test['Path']

arr = []
for files in path:
    image = io.imread(files)
    #resizing the image to match a certain type
    img_resize = transform.resize(image, output_shape=(32,32,3))
    img_reshape = np.reshape(img_resize, (32*32*3))
    arr.append(img_reshape/255.)

x_test = np.array(arr)
print(x_test)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#We can use joblib to save a model using dump

joblib.dump(model, '/content/drive/MyDrive/linear_regressor.pkl')
```

```python
#We can also load the model from joblib using load
joblib.load('/content/drive/MyDrive/linear_regressor.pkl')
```

```python
y_predict = model.predict(x_test)
```

```python
#converting 2D array of vector to 1D array of vector
y_train = np.array(y_train)
y_train
```

```python
y_test = np.array(y_test)
y_test #converting 2D array of vector to 1D array of vector
```

```python
print("Regression Coefficients: ", model.coef_)
```

```python
MSE = mean_squared_error(y_test, y_predict)
R2 = r2_score(y_test, y_predict)
```

```python
print("MSE: ", MSE)
print("R2 error: ", R2)
print('F1 score is: ', model.score(x_train, y_train))
```
